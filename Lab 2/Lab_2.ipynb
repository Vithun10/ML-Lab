```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

aqi = pd.read_csv("city_day.csv")
crop = pd.read_csv("crop_production.csv")


aqi["Date"] = pd.to_datetime(aqi["Date"])

aqi["Year"] = aqi["Date"].dt.year

yearly_aqi = (
    aqi.groupby("Year")["AQI"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10,5))

sns.lineplot(
    data=yearly_aqi,
    x="Year",
    y="AQI",
    marker="o"
)

plt.title("Average AQI by Year")
plt.ylabel("Average AQI")
plt.show()

print(yearly_aqi)
```

Lab 2 - Task 7

```python
aqi["Month"] = aqi["Date"].dt.month

monthly_aqi = (
    aqi.groupby("Month")["AQI"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10,5))

sns.lineplot(
    data=monthly_aqi,
    x="Month",
    y="AQI",
    marker="o"
)

plt.title("Monthly AQI Pattern")
plt.ylabel("Average AQI")
plt.show()

print(monthly_aqi)
```

```python
numeric_cols = aqi.select_dtypes(
    include=np.number
)

corr = numeric_cols.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```

Our analysis indicates that air pollution levels vary considerably across
Indian cities and exhibit clear seasonal patterns. Pollution levels tend to
increase during certain months, particularly around the post-monsoon and
winter periods, when atmospheric conditions trap pollutants closer to the
ground.

The data also shows substantial differences between locations, suggesting
that pollution is not evenly distributed across the country. A small number

Long-term trend analysis suggests that air quality has fluctuated over the
years rather than improving consistently. While some periods show
improvement, others indicate worsening conditions.

Recommendation:
Prioritize targeted pollution-control measures in identified hotspot regions,
especially during high-risk seasonal periods.

Limitation:
The available datasets do not contain sufficient common geographic
information to directly establish whether higher pollution causes lower crop
production. Other factors such as rainfall, irrigation, soil quality and
farming practices are not included and may significantly affect crop yields.


CONCLUSION

The exploratory analysis provided valuable insights into air quality patterns across India. Trend analysis showed that pollution levels have varied over time rather than exhibiting a consistent improvement or decline. Seasonal analysis indicated that AQI levels fluctuate throughout the year, with certain months experiencing noticeably higher pollution levels, supporting the possibility of seasonal environmental influences. Correlation analysis among numerical variables revealed meaningful relationships between several pollutants, suggesting common sources or atmospheric interactions. However, due to the absence of state-level information in the AQI dataset, a direct and reliable assessment of the relationship between air pollution and agricultural production could not be established. Therefore, while the analysis highlights important pollution trends and patterns, additional integrated data would be required to conclusively determine the impact of air quality on crop yields.
